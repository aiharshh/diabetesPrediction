```python
# separating data and labels
x = diabetes_dataset.drop(columns='Outcome', axis=1)
y = diabetes_dataset['Outcome']
```

```python
print(x)
y
```

## Standardization of Data - Data Standardization

```python
# BP can be around 60-70
# BMI can be around 26-36
# Glucose can be 90-130

# for standaridizing our data we have imported StandardScalar from sklearn.preprocessing
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(x)
```

```python
# Now transforming the data
standardized_data = scalar.transform(x)
# what it does is, it takes the data nd removes all the inconsistent values and it keeps all the values in a common range
```

```python
# standardised_data = scalar.fit_transform(x)
# This fits the data and does the standardization in a single step
```

```python
print(standardized_data)
```

```python
# Now all the data have been in the similar range
```

```python
x = standardized_data
y = diabetes_dataset['Outcome']
# now we have have all the data in x and all the data in y
print(x)
print(y)
```

```python
x.shape
```

```python
# Now we need to split our data into training data and testing data
```

## Train Test Split


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify = y, random_state=2)
```

SO here what have done is -
we have x = data above right, so what this train_test_split() will do is, it will split the data into two parts - \
training data - x_train \
testing data - x_test \
our model will be trained on the x_train and the once the model is train we will evaluate our model on the test data\
here the test_size = 0.2, it means 20% of the data is given to evalute the data\
y_train is the label for the x_train data

```python
print(x.shape, x_train.shape, x_test.shape)
```

Training the Model

```python
# SVC = Support Vector Classifier
classifier = svm.SVC(kernel='linear')
#now this will load the svm model in the variable named classifier
# Now we will fit our traning data in this classifier
```

```python
classifier.fit(x_train, y_train) # data and label
```

Evaluation of out model\
Evaluation is basically to check how many times our model is predicting it correctly.

```python
# Model Evaluation
# so first finding the accuracy score on the training data
# and we will predict the labels for it and then we will compare our model prediction and our actual values
# this is how we are going to get the accuracy score muehehehehe
# accuracy score on the training data

x_train_prediction = classifier.predict(x_train)
training_accuracy = accuracy_score(x_train_prediction, y_train)
print("Accuracy score of training data : ",training_accuracy)
```

```python
# accuracy score of the test data
x_test_prediction = classifier.predict(x_test)
testing_accuracy = accuracy_score(x_test_prediction, y_test)
print("Accuracy of the testing data : ", testing_accuracy)
```

```python
# Making a predictive system
input_data = (6,148,72,35,0,33.6,0.627,50)
# changing the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)
# reshape the data as we are predicting are one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
```

## Why did we reshape?
Our model is trained on 768 examples and there are 8 columns\
and as an input we are only giving one data point, and our model expects the input of all 768 values\
This reshape() tell the model that we are only giving one datapoint as input and we only need output for only one datapoint.

```python
# there is one more problem, we cannot give the values like our input data to the input_data.
# because while training the data we standardised the data and
# hence we have to standardise the input data aswell otherwise our model won't be able to make the prediction properly.
```

```python
input_data_standardized = scalar.transform(input_data_reshaped)
print(input_data_standardized)
```

```python
prediction = classifier.predict(input_data_standardized)
print(prediction)
```

```python
if(prediction[0]==1): # we did [0] because prediction is a list
  print("Person is Diabetic")
else:
  print("Person is not Diabetic")
```
